## Explore fastText model

```
from gensim.models import KeyedVectors
model = KeyedVectors.load_word2vec_format(os.path.join(path, 'fastText_model.vec'))
```

```
model.wv.most_similar(positive=["청년"])
```

```
model.wv.most_similar(positive=["청소년"])
```

```
model.wv.most_similar(positive=["강원도"])
```

```
model.wv.similarity("청년", '청소년')
```

```
model.wv.similarity("강원도", '도내')
```

```
model.wv.similarity("강원도", '원주')
```

```
model.wv.similarity('서울', '춘천')
```
